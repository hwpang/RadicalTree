# Get uncertainty for HBI data

```python
import pandas as pd
import numpy as np

from parameters import columns
```

```python
hbi_df = pd.read_csv("../data/hbi.csv")
hbi_df
```

```python
radical_resonance_df = pd.read_csv("../data/radical_resonance.csv")
radical_resonance_df
```

# Merge table

```python
temp_radical_resonance_df = radical_resonance_df.rename(columns={"smiles": "radical_resonance_smiles"})
hbi_unc_df = hbi_df.merge(temp_radical_resonance_df[["radical_resonance_smiles", "num_rotatable_bonds"]], on="radical_resonance_smiles", how="left")
hbi_unc_df = hbi_unc_df.rename(columns={"num_rotatable_bonds": "radical_resonance_num_rotatable_bonds"})
hbi_unc_df
```

# Add representative uncertainty

```python
hbi_unc_df.loc[hbi_unc_df["radical_level_of_theory"]=="CBS-QB3", "unc_radical_H298 (kcal/mol)"] = 1.2
hbi_unc_df.loc[hbi_unc_df["radical_level_of_theory"]=="G4", "unc_radical_H298 (kcal/mol)"] = 1.0

temp_columns = ["unc_radical_" + column for column in columns[1:]]

# uncertainty based on number of rotatable bonds
# for column in temp_columns:
#     hbi_unc_df[column] = hbi_unc_df["radical_resonance_num_rotatable_bonds"].values * 0.5 + 0.5

hbi_unc_df[temp_columns] = 1.0

hbi_unc_df
```

```python
temp_sources = [
    "Thermo library: ../data/dong_pio_liang.py",
    "Thermo library: ../data/pang.py",
    "Thermo library: ../data/johnson_cbs-qb3.py",
]
hbi_unc_df.loc[hbi_unc_df["closed_shell_thermo_source"].isin(temp_sources), "unc_closed_shell_H298 (kcal/mol)"] = 1.2
hbi_unc_df.loc[hbi_unc_df["closed_shell_thermo_source"]=="Thermo library: ../data/johnson_g4.py", "unc_closed_shell_H298 (kcal/mol)"] = 1.0
hbi_unc_df.loc[hbi_unc_df["closed_shell_thermo_source"]=="GAV", "unc_closed_shell_H298 (kcal/mol)"] = 4.0

temp_columns = ["unc_closed_shell_" + column for column in columns[1:]]

# uncertainty based on number of rotatable bonds
# for column in temp_columns:
#     query = hbi_unc_df["closed_shell_thermo_source"]!="GAV"
#     hbi_unc_df.loc[query, column] = hbi_unc_df.loc[query, "radical_resonance_num_rotatable_bonds"].values * 0.5 + 0.5
#     query = hbi_unc_df["closed_shell_thermo_source"]=="GAV"
#     hbi_unc_df.loc[query, column] = hbi_unc_df.loc[query, "radical_resonance_num_rotatable_bonds"].values * 1.0 + 1.0
    
query = hbi_unc_df["closed_shell_thermo_source"]!="GAV"
hbi_unc_df.loc[query, temp_columns] = 1.0
query = hbi_unc_df["closed_shell_thermo_source"]=="GAV"
hbi_unc_df.loc[query, temp_columns] = 2.0

hbi_unc_df
```

```python
for column in columns:
    hbi_unc_df["unc_HBI_" + column] = hbi_unc_df["unc_radical_" + column] + hbi_unc_df["unc_closed_shell_" + column]
    
hbi_unc_df
```

```python
hbi_unc_df.to_csv("../data/hbi_unc.csv", index=False)
```
